## Demo 3: Import color naming data
    
> Each of the 330 color chips was named by speakers of 110 different languages.

______________________________________________

Load naming data. 

`namingData` is a hierarchical dictionary organized as follows:

**language _(1 - 110)_ &rarr; speaker _(1 - *range varies per language*)_ &rarr; chip index _(1 - 330)_ &rarr; color term**

```python
namingData = readNamingData('term.txt')
```

For example, to obtain naming data from language 1 and speaker 1 for all 330 color chips:

```python
#namingData[1]
```

```python
naming_df = pd.DataFrame.from_dict(namingData, orient= 'index').reset_index()
naming_df = naming_df.rename(columns = {'index':'Language'})
naming_df.head(4)
#naming_df = naming_df.fillna(0)
```

```python
#the count includes the 1st column 
naming_df.count(axis =1)
```

```python
#so the actual sum of the total speaker would be  
sum(naming_df.count(axis =1))-110
```

```python
sp_of_each_lang = [] # includes corresponding language values with the colour codes 

for index, rows in naming_df.iterrows():
    #print(type(rows))
    row = pd.DataFrame(rows).dropna()
    for i, r in row.iterrows():
        #print(r) 
        sp_of_each_lang.append(r.iloc[0])
        
    #print("---{}---".format(index))
```

```python
sp_of_each_lang[:4]
```

```python
len(sp_of_each_lang) 
```

```python
sp_of_each_lg = [] # only includes the colour code doesnt include corresponding language values
unique = []

for i in range(len(sp_of_each_lang)): 
    if isinstance(sp_of_each_lang[i], dict):
        #print(i)
        il = list(sp_of_each_lang[i].values())
        unqe = len(set(il))
        sp_of_each_lg.append(il) 
        unique.append(unqe)
        

print(sp_of_each_lang[-1])
print(unique[:10])
```

```python
#sp_of_each_lg is a list of list each list element includes 310 color responses  
print(len(sp_of_each_lg))
print(len(unique))
```

<H4> Plotting Gender and Unique Color names on a scatter plot </H4>

```python
# Create two lists = one containing all the gender, two containing all the unique color numbers, three containing their ages
one = []
two = []
three = []
for vals in profile:
    if vals[0][0][1] == 'M' or vals[0][0][1] == 'X':
        one.append('M')
        two.append(vals[1])
        three.append(vals[0][0][0])
    elif vals[0][0][1] == 'F' or vals[0][0][1] == 'f':
        one.append('F')
        two.append(vals[1])
        three.append(vals[0][0][0])
# Make a dataframe from these three lists for easy plotting purposes
gender_age_df = pd.DataFrame(list(zip(one, two, three)), columns = ['gender', 'unique', 'age'])

# Add language number from the old variable gd to this dataframe
gender_age_df
```

```python
sns.scatterplot(x='gender', y = 'unique', data=gender_age_df)
```

<H4> Plotting Age and Unique Color names on a scatter plot </H4>

```python
# Creating a new variable with the age column sorted smallest to biggest
age_sorted = gender_age_df.sort_values(by=['age'])
```

```python
# Removing all rows with age < 10 years old cause these are errors
age_sorted_new = age_sorted[(age_sorted['age'].astype('int64')) >= 10]

plt.figure(figsize = (20,15))
plt.plot(age_sorted_new['age'], age_sorted_new['unique'], 'bo')
```

```python
# lab 3 110 languages -> n
threshold = 30
age_sorted_young = age_sorted_new[(age_sorted_new['age'].astype('int64')) <= threshold]
age_sorted_young_female = age_sorted_young[age_sorted_young['gender'] == 'F']
age_sorted_young_male = age_sorted_young[age_sorted_young['gender'] == 'M']

age_sorted_old = age_sorted_new[(age_sorted_new['age'].astype('int64')) > threshold]
age_sorted_old_female = age_sorted_old[age_sorted_old['gender'] == 'F']
age_sorted_old_male = age_sorted_old[age_sorted_old['gender'] == 'M']

# Changing gender term for old females from 'F' to 'old_F'
age_sorted_old_female['gender'] = 'old_F'
```

```python
# threshold analysis
stat, p = stats.ttest_ind(age_sorted_young_male['unique'], age_sorted_young_female['unique'])
print("t-test result between young males and females. T-stat = ", stat, " and associated p-value is = ", p)

stat, p = stats.ttest_ind(age_sorted_old_male['unique'], age_sorted_young_female['unique'])
print("t-test result between old males and females. T-stat = ", stat, " and associated p-value is = ", p)

stat, p = stats.ttest_ind(age_sorted_old_female['unique'], age_sorted_young_female['unique'])
print("t-test result between young females and old females. T-stat = ", stat, " and associated p-value is = ", p)

stat, p = stats.ttest_1samp(age_sorted_young_male['unique'], age_sorted_young_female['unique'])
print("t-test result between young males and females. T-stat = ", stat, " and associated p-value is = ", p)
```
